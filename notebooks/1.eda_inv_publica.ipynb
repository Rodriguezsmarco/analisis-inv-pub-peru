## 1. IMPORTACIÓN DE LIBRERÍAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
pd.options.display.float_format = '{:.2f}'.format
```

## 2. CARGA DE DATASET (DATA)

```python
data = pd.read_csv("../data/DETALLE_INVERSIONES.csv")
```

3.1 VALORES FALTANTES

```python
faltantes = data.isnull().sum().sort_values(ascending=False)
faltantes[faltantes > 0]
```

```python
data.describe()
```

3.3 CANTIDAD DE PROYECTOS POR NIVEL DE GOBIERNO

```python
data["NIVEL"].value_counts().plot(kind="bar", figsize=(8,5))

plt.title("Niveles de gobierno con más proyectos")
plt.show()
```

3.4 CANTIDAD DE PROYECTOS POR REGIÓN

```python
data["DEPARTAMENTO"].value_counts().sort_values(ascending=True).plot(
    kind="barh",
    figsize=(10,8)
)

plt.title("Regiones con más proyectos")
plt.show()
```

3.5 MONTO DE INVERSIÓN POR SECTOR

```python
data.groupby("SECTOR")["COSTO_ACTUALIZADO"].sum().sort_values().tail(12).plot(
    kind="barh",
    figsize=(10,8)
)

plt.title("Los 10 sectores con más inversión")
plt.show()
```

3.6 DISTRIBUCIÓN DEL AVANCE DE EJECUCIÓN

```python
plt.figure(figsize=(8,5))

sns.histplot(data["AVANCE_EJECUCION"], bins=50)

plt.title("Distribución del avance de ejecución")
plt.show()
```

3.7 AVANCE PROMEDIO POR NIVEL DE GOBIERNO

```python
data.groupby("NIVEL")["AVANCE_EJECUCION"].mean().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Avance promedio por nivel de gobierno")
plt.show()
```

3.8 OUTLIERS DE INVERSIÓN

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=data["COSTO_ACTUALIZADO"])

plt.title("Outliers inversión pública")
plt.show()
```
