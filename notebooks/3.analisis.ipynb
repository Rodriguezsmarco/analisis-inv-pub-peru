## 8. ANÁLISIS DE TIEMPO

8.1 DEFINICIÓN DE VARIABLES

```python
dat["FECHA_VIABILIDAD"] = pd.to_datetime(dat["FECHA_VIABILIDAD"], errors="coerce")
dat["ANIO_VIABILIDAD"] = dat["FECHA_VIABILIDAD"].dt.year
dat["FEC_INI_EJECUCION"] = pd.to_datetime(dat["FEC_INI_EJECUCION"], errors="coerce")
dat["ANIO_EJECUCION"] = dat["FEC_INI_EJECUCION"].dt.year
```

8.2 DEFINICIÓN DE PROCEDIMIENTO

```python
def analisis_tiempo(data, variable):
    """Ejecuta el análisis respecto de inversión, ejecución, cantidad, sobrecostos y riesgo financiero basado en una variable temporal específica."""
    print(f"Generando reporte basado en: {variable}")
        
    metricas = {
        "COSTO_ACTUALIZADO": ("sum", "Evolución de la Inversión Pública", "Monto Total"),
        "AVANCE_EJECUCION": ("mean", "Ejecución Promedio", "Avance (%)"),
        "SOBRECOSTO": ("mean", "Sobrecosto Promedio", "Monto Sobrecosto"),
        "RIESGO_FINANCIERO": ("mean", "Riesgo Promedio", "Nivel de Riesgo")
    }
        
    for col, (metodo, titulo, ylabel) in metricas.items():        
        serie = data.groupby(variable)[col].agg(metodo)        
        plt.figure(figsize=(10, 5))
        serie.plot(marker="o")
        plt.title(f"{titulo} por {variable}")
        plt.ylabel(ylabel)
        plt.xlabel(variable)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()
           
    proyectos = data[variable].value_counts().sort_index()
    plt.figure(figsize=(10, 5))
    proyectos.plot(marker="o")
    plt.title(f"Número de Proyectos por {variable}")
    plt.ylabel("Cantidad")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
    
    resumen = data.groupby(variable).agg({
        "COSTO_ACTUALIZADO": "sum",
        "AVANCE_EJECUCION": "mean",
        "SOBRECOSTO": "mean",
        "RIESGO_FINANCIERO": "mean"
    })
  
    return resumen
```

8.3 ANALISIS SEGÚN SU VIABILIDAD

```python
analisis_tiempo(dat, "ANIO_VIABILIDAD")
```

8.4 ANÁLISIS SEGÚN SU EJECUCIÓN

```python
analisis_tiempo(dat, "ANIO_EJECUCION")
```

11.4 CURVA DE CONCENTRACIÓN DE PROYECTOS

```python
sorted_cost = np.sort(dat["COSTO_ACTUALIZADO"])[::-1]

cum_cost = np.cumsum(sorted_cost)
cum_cost = cum_cost / cum_cost[-1]

plt.figure(figsize=(8,5))
plt.plot(cum_cost)
plt.title("Curva de Concentración de Inversión")
plt.xlabel("Proyectos acumulados")
plt.ylabel("Inversión acumulada")
plt.show()
```
